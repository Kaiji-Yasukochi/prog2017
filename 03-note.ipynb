```python
# coding:utf-8
print(' こんにちは')
#身長の入力
h = input(' 身長（m）？ ')
print(' 身長は{0}m ですね'.format(h))
#体重の入力
w = input(' 体重（kg）？ ')
10
print(' 体重は{0}kg ですね'.format(w))
#BMI の計算
bmi = float(w) / float(h) ** 2
#BMI の表示
print('BMI は{0}です'.format(bmi))
print(' 約{0:.1f}です'.format(bmi))
```
